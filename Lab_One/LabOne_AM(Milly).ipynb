Definign a function to calculate the area of a rectangle. User will input the width and length which will return the output in cm2.

```python
def rectarea(height,width):
  return height*width

h = float(input("Enter the height of your rectangle in cm "))
w = float(input("Enter the width of your rectangle in cm "))

out = rectarea(h,w)
print ("The area of rectangle is {}cm2". format(out))
```

Using f string to shorten the width of the decimal places.

```python
def rectarea(height,width):
  return height*width

h = float(input("Enter the height of your rectangle in cm "))
w = float(input("Enter the width of your rectangle in cm "))

out = rectarea(h,w)
print (f'The area of the rectangle is {out:.2f}cm2')
```
